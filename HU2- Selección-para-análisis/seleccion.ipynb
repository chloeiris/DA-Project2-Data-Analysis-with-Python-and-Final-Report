Aunque por petición del cliente tenemos un dataframe total del que podríamos seleccionar las variables, como equipo hemos decidido repartirnos la tarea de selección repartiéndonos un archivo original a cada una. En este notebook haremos la unión de esa selección. Vamos a unir 3/4 archivos, dado que en uno de ellos no habia columnas de nuestro interés.

```python
import pandas as pd
import numpy as np
```

# Carga de los datasets con las variables seleccionadas

```python
kaggel_final= pd.read_csv('data/variables-seleccionadas/kaggel_modificado2.csv', index_col=0)
kaggel_final.head(2)
```

```python
txt_final= pd.read_csv('data/variables-seleccionadas/txt_csv2.csv', index_col=0)
txt_final.head(2)
```

```python
xml_final= pd.read_csv('data/variables-seleccionadas/xml_csv_2.csv', index_col=0)
xml_final.head(2)
```

```python
# Renombramos las columnas de uno de los dataframes antes de la unión
nuevas_columnas = {col: col.lower().replace(" ", "_") for col in kaggel_final.columns}
kaggel_final.rename(columns = nuevas_columnas, inplace = True)
kaggel_final.head(1)

```

# Unión de las variables seleccionadas

```python
data_final = pd.concat([xml_final, txt_final, kaggel_final], axis=1)
data_final.head(2)
```

# Guardado en csv

```python
data_final.to_csv('data_seleccionada.csv')
```
