```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
pd.options.display.max_columns = None
plt.rcParams["figure.figsize"] = (30,25) 
```

```python
# Cargo el dataframe con fichero del total seleccionado
df = pd.read_csv('data_seleccionada.csv', index_col = 0)
```

# EDA

```python
# Explorando el dataset
df.shape
```

```python
df.describe(include= 'all').T
```

```python
## Columnas por las que voy a filtrar [q40] y ['q_5_role']
# Como la q 40 esta repartida en varias columnas, voy a hacer un dataframe solo con esas para analizarlo mejor
df40 = df.loc[: , ['q40_part_1','q40_part_2','q40_part_3','q40_part_4','q40_part_5','q40_part_6','q40_part_7','q40_part_8','q40_part_9','q40_part_10','q40_part_11']]
df40.head()
```

```python
df40.shape
```

```python
df40.describe().T

# Aqui veo la formación que corresponde a cada pregunta, y en 'count' su frecuencia
```

```python
# Con df.index y df.values podría extraer los nombres para renombrar las columnas del dataset
df40.describe().T['top'].values  
```

```python
# Vamos a ordenarlos en orden de popularidad pues a simple vista se ve que coursera es la más frecuente

df_count_courses = df40.describe().T.loc[:, ['count', 'top']]
df_count_courses.sort_values(by= 'count', ascending = False)

# Otra forma de hacerlo
# describe().T.groupby('top')#['count']#.sort_values(ascending = False)
```

```python
# Haremos una gráfica rápida de barras para ver la frecuencia
fig, ax = plt.subplots(figsize = (18, 12))
ax.bar(x = df_count_courses['top'], height = df_count_courses['count'] )
plt.show()
```

```python
# Quiero hace un gráfico con hue por "role", por lo que tengo que hacer un df con la q5 también

df40_5 = df.loc[: , ['q_5_role','q40_part_1','q40_part_2','q40_part_3','q40_part_4','q40_part_5','q40_part_6','q40_part_7','q40_part_8','q40_part_9','q40_part_10','q40_part_11']]
df40_5.head()
```

```python
df40_5.describe().T
```

```python
# Hacemos una tabla de frecuencias usando group by

frecuencia_courses = df40_5.groupby('q_5_role').count()
frecuencia_courses
```

```python
# Vamos a renombrar la tabla de frecuencias para poder graficarla mejor
nombres = df40.describe().T['top'].values
nombres_nuevos_list = list(nombres)
```

```python
nombres_antiguos_list = frecuencia_courses.columns
```

```python
rename_dict = dict(zip(nombres_antiguos_list, nombres_nuevos_list))
rename_dict
```

```python
frecuencia_courses.rename(columns =rename_dict, inplace=True)
```

```python
# Tabla de frecuencias renombrada
frecuencia_courses
```

```python
# Vamos a plotear las frecuencias de cada curso según el rol profesional
plt.rcParams["figure.figsize"] = (20,15) 
frecuencia_courses.plot.bar();
```

```python
# Voy a filtrar solo para los cursos que consumen los ML engineers

cursos_elegidos_ml_engineer = pd.DataFrame(frecuencia_courses.loc['Machine Learning Engineer', :].sort_values())
cursos_elegidos_ml_engineer
```

```python
plt.rcParams["figure.figsize"] = (9, 6) 
cursos_elegidos_ml_engineer.plot.bar();
#plt.xticks(rotation=45, horizontalalignment= 'right');
```
